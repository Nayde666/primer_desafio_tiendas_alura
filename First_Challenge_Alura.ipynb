## Importación de datos

## 2.Ventas por categoría

```python
def categorias(num_tienda):
  categorias = []
  for i in num_tienda['Categoría del Producto']:
    if i not in categorias:
      categorias.append(i)
  return categorias

categorias_t1 = categorias(tienda)
categorias_t2 = categorias(tienda2)
categorias_t3 = categorias(tienda3)
categorias_t4 = categorias(tienda4)

print(categorias_t1)
print(categorias_t2)
print(categorias_t3)
print(categorias_t4)
```

```python
def contar_categorias(num_tienda):
    conteo = {}
    for categoria in num_tienda['Categoría del Producto']:
        if categoria in conteo:
            conteo[categoria] += 1
        else:
            conteo[categoria] = 1
    return conteo

tiendas = [tienda, tienda2, tienda3, tienda4]

for i, t in enumerate(tiendas, start=1):
  print(f'Los conteos por categoria para la tienda {i} son: {contar_categorias(t)}')
```

```python
# Categoria más y menos vendida
def categoria_mas_min(conteo):
  mas_vendido = max(conteo, key=conteo.get)
  menos_vendido = min(conteo, key=conteo.get)
  return mas_vendido, menos_vendido


for i, t in enumerate(tiendas, start=1):
    conteo = contar_categorias(t)
    mas, menos = categoria_mas_min(conteo)
    print(f"Tienda {i}:")
    print(f"  Categoria más vendida: {mas} ({conteo[mas]} veces)")
    print(f"  Categoria menos vendida: {menos} ({conteo[menos]} veces)\n")
```

```python
conteo = contar_categorias(tienda)  # cambiar segun tienda

categorias = list(conteo.keys())
cantidades = list(conteo.values())

plt.figure(figsize=(5, 5))
plt.pie(
    cantidades,
    labels=categorias,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors
)
plt.title('Distribución por Categoría - Tienda 1')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

## 3. Calificación promedio de la tienda

```python
def calificacion_promedio(num_tienda):
  suma = num_tienda['Calificación'].sum()
  total_calificaciones = num_tienda['Calificación'].count()
  promedio = suma / total_calificaciones
  return round(promedio,2)

for i, t in enumerate(tiendas, start=1):
    print(f'El promedio para la tienda {i} es: {calificacion_promedio(t)}')
```

## Productos más y menos vendidos

```python
def conteo_productos(num_tienda):
    conteo = {}
    for producto in num_tienda['Producto']:
        if producto in conteo:
            conteo[producto] += 1
        else:
            conteo[producto] = 1
    return conteo

for i, t in enumerate(tiendas, start=1):
    print(f'El conteo de productos para la tienda {i} es: {conteo_productos(t)}')
```

```python
def producto_mas_min(conteo):
  mas_vendido = max(conteo, key=conteo.get)
  menos_vendido = min(conteo, key=conteo.get)
  return mas_vendido, menos_vendido


for i, t in enumerate(tiendas, start=1):
    conteo = conteo_productos(t)
    mas, menos = producto_mas_min(conteo)
    print(f"Tienda {i}:")
    print(f"  Más vendido: {mas} ({conteo[mas]} veces)")
    print(f"  Menos vendido: {menos} ({conteo[menos]} veces)\n")
```

```python
def conteo_productos(num_tienda):
    conteo = {}
    for producto in num_tienda['Producto']:
        if producto in conteo:
            conteo[producto] += 1
        else:
            conteo[producto] = 1
    return conteo

def producto_mas_min(conteo):
    mas_vendido = max(conteo, key=conteo.get)
    menos_vendido = min(conteo, key=conteo.get)
    return mas_vendido, conteo[mas_vendido], menos_vendido, conteo[menos_vendido]

tiendas = [tienda, tienda2, tienda3, tienda4]

labels_tiendas = []
productos_mas = []
cantidades_mas = []
productos_menos = []
cantidades_menos = []

for i, t in enumerate(tiendas, start=1):
    conteo = conteo_productos(t)
    mas, cant_mas, menos, cant_menos = producto_mas_min(conteo)

    labels_tiendas.append(f'Tienda {i}')
    productos_mas.append(mas)
    cantidades_mas.append(cant_mas)
    productos_menos.append(menos)
    cantidades_menos.append(cant_menos)

x = np.arange(len(labels_tiendas))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
barras_verdes = ax.bar(x - width/2, cantidades_mas, width, label='Más vendido', color='green')
barras_rojas = ax.bar(x + width/2, cantidades_menos, width, label='Menos vendido', color='red')

ax.set_xlabel('Tienda')
ax.set_ylabel('Cantidad vendida')
ax.set_title('Comparación del Producto Más y Menos Vendido por Tienda')
ax.set_xticks(x)
ax.set_xticklabels(labels_tiendas)
ax.legend()

for i in range(len(labels_tiendas)):
    ax.annotate(f'{productos_mas[i]}\n({cantidades_mas[i]})',
                xy=(x[i] - width/2, cantidades_mas[i]),
                xytext=(0, 3), textcoords='offset points',
                ha='center', va='bottom', fontsize=8)

    ax.annotate(f'{productos_menos[i]}\n({cantidades_menos[i]})',
                xy=(x[i] + width/2, cantidades_menos[i]),
                xytext=(0, 3), textcoords='offset points',
                ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()
```

## Envío promedio por tienda

```python
def envio_promedio(num_tienda):
  suma = num_tienda['Costo de envío'].sum()
  total_registros = num_tienda['Costo de envío'].count()
  promedio = suma / total_registros
  return round(promedio,2)

tiendas = [tienda, tienda2, tienda3, tienda4]
for i, t in enumerate(tiendas, start=1):
    print(f'El promedio de envío para la tienda {i} es: $ {envio_promedio(t)}')
```
